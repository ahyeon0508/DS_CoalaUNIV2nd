```python
import pandas as pd

df = pd.read_csv('data/house.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
house_data = df.drop(['id', 'date'], axis=1) # axis=1 여기선 열을 가리킨다
house_data.head()
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(house_data.corr(), annot=True, fmt='.2f', square=True)
# annot=False -> 상관계수 사라짐
plt.show()
```

```python
house_data.corr() # 상관계수 출력
```

```python
sns.catplot(data=house_data, x='bathrooms', y='price', aspect=2)
```

```python
sqft_max = house_data['sqft_living'].max()
```

```python
print('거실면적 최대값 : ', sqft_max)
```

```python
print('집값 최소값', house_data['price'].min())
print('집값 최대값', house_data['price'].max())
print('집값 평균값', house_data['price'].mean())
```
