```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# (1,90) (2,100) (3,50)
plt.plot([1,2,3], [90,100,50])
plt.show()
```

```python
plt.plot(['Math', 'Korean', 'English'], [90, 100, 70])

plt.plot(['Math', 'Korean', 'English'], [70, 100, 70])

plt.xlabel('subjects')
plt.ylabel('score')
plt.title('Midterm')

plt.legend(['choi', 'kim'])

plt.show()
```

```python
x = ['Math', 'Korean', 'English']
y = [90, 100, 50]

plt.xlabel('Subjects')
plt.ylabel('Score')

plt.bar(x, y, width = 0.6, color = '#310424')
```

```python
plt.plot([1, 2, 3], [90, 100, 50], marker='s', markersize=30, linestyle = ":")

plt.grid(True)
plt.axis([1,3,8,100])

plt.show()
```

#  csv  

```python
import pandas as pd
```

```python
train = pd.read_csv('data/train.csv')

train
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
sns.relplot(data=train, x='Age', y = 'Survived')
```

```python
sns.relplot(data=train, x='Age', y = 'Survived', hue='Sex', aspect=4)
```

```python
sns.catplot(data=train, x='Pclass', y='Fare', aspect=4, jitter=False)
```

```python
sns.catplot(data=train, x='Pclass', y='Fare', aspect=4, hue='Sex')
```

```python
sns.pairplot(data=train, hue='Sex', 
             x_vars=['Pclass', 'Age', 'Fare'], y_vars=['Survived', 'Age', 'Fare'], height=3)
```

```python
sns.catplot(data=train, x='Pclass', y='Fare', hue='Sex', kind = 'box')
```
